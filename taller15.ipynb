```python
import numpy as np

 #APLICA EL INVERSO PARA LA MATRIZ A

filas = int(input("Ingrese el número de filas de la matriz: "))
columnas = int(input("Ingrese el número de columnas de la matriz: "))


print("Ingrese los elementos de la matriz, separados por espacios:")
matriz = []
for i in range(filas):
    fila = list(map(float, input().split()))
    matriz.append(fila)

matriz = np.array(matriz)

# Comprobar que la matriz sea cuadrada
if filas != columnas:
    print("La matriz no es cuadrada y, por lo tanto, no tiene una matriz inversa.")
else:

    identidad = np.identity(filas)

    aumentada = np.concatenate((matriz, identidad), axis=1)

     # Metodo Gauss-Jordan
    for i in range(filas):
        # Dividir la fila i por el elemento en la diagonal principal
        diagonal_principal = aumentada[i][i]
        aumentada[i] /= diagonal_principal

      
        for j in range(i + 1, filas):
            factor = aumentada[j][i]
            aumentada[j] -= factor * aumentada[i]

    for i in reversed(range(filas)):
       
        for j in reversed(range(i)):
            factor = aumentada[j][i]
            aumentada[j] -= factor * aumentada[i]

  
    inversa = aumentada[:, filas:]

    # Imprimir la matriz inversa
    print("La matriz inversa de la matriz A es:")
    print(inversa)

import numpy as np

 #APLICA EL INVERSO PARA LA MATRIZ B

filas = int(input("Ingrese el número de filas de la matriz: "))
columnas = int(input("Ingrese el número de columnas de la matriz: "))


print("Ingrese los elementos de la matriz, separados por espacios:")
matriz = []
for i in range(filas):
    fila = list(map(float, input().split()))
    matriz.append(fila)

matriz = np.array(matriz)

# Comprobar que la matriz sea cuadrada
if filas != columnas:
    print("La matriz no es cuadrada y, por lo tanto, no tiene una matriz inversa.")
else:

    identidad = np.identity(filas)

    aumentada = np.concatenate((matriz, identidad), axis=1)

     # Metodo Gauss-Jordan
    for i in range(filas):
        # Dividir la fila i por el elemento en la diagonal principal
        diagonal_principal = aumentada[i][i]
        aumentada[i] /= diagonal_principal

      
        for j in range(i + 1, filas):
            factor = aumentada[j][i]
            aumentada[j] -= factor * aumentada[i]

    for i in reversed(range(filas)):
       
        for j in reversed(range(i)):
            factor = aumentada[j][i]
            aumentada[j] -= factor * aumentada[i]

  
    inversa = aumentada[:, filas:]

    # Imprimir la matriz inversa
    print("La matriz inversa de la matriz B es:")
    print(inversa)

```
