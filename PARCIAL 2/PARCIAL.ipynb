```python
import numpy as np

#PRIMER PUNTO PARCIAL

filas_a = int(input("Introduzca el número de filas de la matriz A: "))
columnas_a = int(input("Introduzca el número de columnas de la matriz A: "))
filas_b = int(input("Introduzca el número de filas de la matriz B: "))
columnas_b = int(input("Introduzca el número de columnas de la matriz B: "))


A = np.zeros((filas_a, columnas_a))
B = np.zeros((filas_b, columnas_b))

print("Introduzca los valores de la matriz A: ")
for i in range(filas_a):
    for j in range(columnas_a):
        A[i,j] = float(input(f"A4[{i},{j}]: "))

print("Introduzca los valores de la matriz B: ")
for i in range(filas_b):
    for j in range(columnas_b):
        B[i,j] = float(input(f"B[{i},{j}]: "))


producto_A_B = np.dot(A, B)


print("B × A =")
print(producto_A_B)
```

```python

#SEGUNDO PUNTO PARCIAL

import numpy as np
import matplotlib.pyplot as plt


x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([1.8, 2, 2.2, 2.6, 2.9, 3.3, 3.7, 4.1])


coeffs_linear = np.polyfit(x, y, 1)
line_func = np.poly1d(coeffs_linear)


plt.scatter(x, y, label='Puntos')
plt.plot(x, line_func(x), label='Aproximación lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Aproximación lineal')
plt.grid(True)
plt.show()
# Ajuste de un polinomio de grado 2
coeffs_poly2 = np.polyfit(x, y, 2)
poly2_func = np.poly1d(coeffs_poly2)

# Crear la gráfica
plt.scatter(x, y, color='red', label='Puntos')
plt.plot(x, poly2_func(x), label='Aproximación G2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Aproximación polinomial de grado 2 de puntos')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

#TERCER PUNTO PARCIAL

x = [1, 2, 3, 4, 5]
f = [1.8, 2, 2.5, 2.8, 3.5]

n = len(x)
h = [x[i+1] - x[i] for i in range(n-1)]
alpha = [0 for i in range(n)]
for i in range(1, n-1):
    alpha[i] = 3 * ((f[i+1] - f[i])/h[i] - (f[i] - f[i-1])/h[i-1])

l = [1] + [0]*(n-2)
mu = [0] * n
z = [0] * n
for i in range(1, n-1):
    l[i] = 2 * (x[i+1] - x[i-1]) - h[i-1] * mu[i-1]
    mu[i] = h[i]/l[i]
    z[i] = (alpha[i] - h[i-1]*z[i-1])/l[i]

l[-1] = 1  
z[-1] = 0
c = [0] * n
b = [0] * n
d = [0] * n
for j in range(n-2, -1, -1):
    c[j] = z[j] - mu[j] * c[j+1]
    b[j] = (f[j+1] - f[j])/h[j] - h[j]*(c[j+1] + 2*c[j])/3
    d[j] = (c[j+1] - c[j])/(3*h[j])

poly = []
for i in range(n-1):
    poly.append([d[i], c[i], b[i], f[i]])

x_est = 3.2
for i in range(n-1):
    if x_est >= x[i] and x_est <= x[i+1]:
        f_est = poly[i][0]*(x_est-x[i])**3 + poly[i][1]*(x_est-x[i])**2 + poly[i][2]*(x_est-x[i]) + poly[i][3]
        break

plt.scatter(x, f)
for i in range(n-1):
    x_vals = [x[i] + j*(x[i+1]-x[i])/100 for j in range(101)]
    y_vals = [poly[i][0]*(x_vals[j]-x[i])**3 + poly[i][1]*(x_vals[j]-x[i])**2 + poly[i][2]*(x_vals[j]-x[i]) + poly[i][3] for j in range(101)]
    plt.plot(x_vals, y_vals)

plt.plot([x_est], [f_est], 'ro')
plt.show()

print(f"f(3.2) = {f_est}")
```
