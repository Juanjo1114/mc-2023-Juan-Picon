```python
import numpy as np
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5, 6, 7]
y = [3.5, 2.5, 3, 1.5, 2, 1.3, 1, 0.3]


mean_x = np.mean(x)
mean_y = np.mean(y)


var_x = np.var(x)


cov_xy = np.cov(x, y, ddof=0)[0][1]

# Calcular la pendiente y la intersección de la línea recta
m = cov_xy / var_x
b = mean_y - m * mean_x

x_range = np.arange(min(x), max(x) + 1, 0.1)

# Aquí se calcula la pendiente
y_range = m * x_range + b

#Se grafican los puntos (x, y)
plt.scatter(x, y)
plt.plot(x_range, y_range, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresión lineal por mínimos cuadrados')
plt.show()
```
